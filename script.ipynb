## **Dividindo o Dataframe principal por tabelas do banco**

### **Tabela escola**

```python
# df dados escola
escola = df[['co_municipio_esc', 'no_municipio_esc', 'co_uf_esc', 'sg_uf_esc',
       'tp_dependencia_adm_esc', 'tp_localizacao_esc', 'tp_sit_func_esc']]
```

```python
escola.isnull().sum()
```

```python
escola = escola.dropna()
```

```python
escola.dtypes
```

```python
len(escola)
```

```python
escola.head()
```

```python
# realizar o cast das colunas
escola.loc[:,"co_municipio_esc"] = escola["co_municipio_esc"].astype("Int64")
escola.loc[:,"co_uf_esc"] = escola["co_uf_esc"].astype("Int64")
escola.loc[:,"tp_dependencia_adm_esc"] = escola["tp_dependencia_adm_esc"].astype("Int64")
escola.loc[:,"tp_localizacao_esc"] = escola["tp_localizacao_esc"].astype("Int64")
escola.loc[:,"tp_sit_func_esc"] = escola["tp_sit_func_esc"].astype("Int64")
```

```python
escola.head()
```

```python
escola = escola.drop_duplicates(subset=["co_municipio_esc"])
```

```python
# Mount SQL Values according count the columns
qtdColumn = len(escola.columns)
```

```python
columns = (escola.columns).tolist()
```

```python
# Mount tuple for insert
escola = tuple(x for x in escola.to_numpy())
```

```python
# Connect to the database
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='teste123',
                             database='mysql',
                             cursorclass=pymysql.cursors.DictCursor)

with connection:
    with connection.cursor() as cursor:
        # Create a new record
        for item in escola:
            item = tuple(item)
            sql = f'''INSERT INTO enem.ESCOLA (CO_MUNICIPIO_ESC,
                        NO_MUNICIPIO_ESC,
                        CO_UF_ESC,
                        SG_UF_ESC,
                        TP_DEPENDENCIA_ADM_ESC,
                        TP_LOCALIZACAO_ESC,
                        TP_SIT_FUNC_ESC) VALUES {item}'''
            cursor.execute(sql)
            connection.commit()
            
result = cursor.fetchone()
print(result)
```
